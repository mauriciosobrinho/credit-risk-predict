### Text Processing
+ Reduce Noise
    - Punct
    - Special Characteres
    - Stop Words
+ Lemmantization/Stemming

```python
dir(ntx)
```

```python
# Remove stopwords
df["clean_items"] = df["items"].apply(ntx.remove_stopwords)
df [["clean_items", "items"]]
```

```python
# Remove special characters
df["clean_items"] = df["clean_items"].apply(ntx.remove_special_characters)
df [["clean_items", "items"]]
```

```python
# Reduce to lowercase
df["clean_items"] = df["clean_items"].str.lower()
df [["clean_items", "items"]]
```

### Buiding Features From The Text
+ Convert words to vector of number
+ Tfidf matrix
+ Count
+ Hash vector

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
Xfeatures = df["clean_items"]
yLabels = df["over5_D23"]
```

```python
Xfeatures.head()
```

```python
tfidf_vec = TfidfVectorizer()
X = tfidf_vec.fit_transform(Xfeatures)
```

```python
# Convert to DataFrame
df_vec = pd.DataFrame(X.todense(), columns=tfidf_vec.get_feature_names())
```
